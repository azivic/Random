```
image2=Image.open("math.jpg")
plt.imshow(image2)
```

```
#transposing
image2.transpose(Image.FLIP_TOP_BOTTOM)
```

```
image2.transpose(Image.ROTATE_270)
```

```
#resize-the best resolution is when you use bicubic and lanczos interpolation algorithms
size=(500,300)
size1=(380,200)
plt.figure(figsize=(20,30))
resized_image_bic=image2.resize(size, Image.BICUBIC)
resized_image_lan=image2.resize(size1, Image.LANCZOS)

plt.subplot(3,2,1)
plt.imshow(resized_image_bic)
plt.title("Bicubic")

plt.subplot(3,2,2)
plt.imshow(resized_image_lan)
plt.title("Lanczos")
```

```
#incresing the image quality
from PIL import ImageEnhance as ie

plt.figure(figsize=(15, 20))

br=ie.Brightness(resized_image_lan).enhance(1.2)
col=ie.Color(resized_image_lan).enhance(3)
con=ie.Contrast(resized_image_lan).enhance(2.5)
sh=ie.Sharpness(resized_image_lan).enhance(2)

plt.subplot(4,2,1)
plt.imshow(br)
plt.title("Brightness")

plt.subplot(4,2,2)
plt.imshow(sh)
plt.title("Sharpness")

plt.subplot(4,2,3)
plt.imshow(col)
plt.title("Color")

plt.subplot(4,2,4)
plt.imshow(con)
plt.title("Contrast")
```
